```python
Handling missing data
```

```python
df[(df['Item'] == 'UNKNOWN')& (df['Quantity'] == 'UNKNOWN' )]
```

```python
drop=df[(df['Item'] == 'UNKNOWN')& (df['Quantity'] == 'UNKNOWN' )]
drop
```

```python
df.shape
```

```python
df = df[~df['Transaction ID'].isin(drop['Transaction ID'])]
# drop all row from the data
```

```python
df['Item'].unique()
```

```python
import numpy as np
df.replace(['UNKNOWN','ERROR'],np.nan,inplace=True)
```

```python
df['Item'].fillna(df['Item'].mode()[0],inplace=True)
```

```python
df['Item'].unique()
```

```python
df
```

```python
df['Quantity'].unique()
```

```python
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
```

```python
df['Quantity'].fillna(df['Quantity'].median(),inplace=True)
```

```python
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df['Price Per Unit'].fillna(df['Price Per Unit'].median(),inplace=True)
```

```python
df['Price Per Unit'].unique()
```

```python
df['Total Spent'].unique()
```

```python
df['Total Spent']=df['Price Per Unit']*df['Quantity']
```

```python
df['Total Spent'].unique()
```

```python
df.info()
```

```python
df['Payment Method'].unique()
```

```python
df['Payment Method'].fillna(df['Payment Method'].mode()[0],inplace=True)
```

```python
df['Payment Method'].unique()
```

```python
df['Location'].unique()
```

```python
df['Location']=df['Location'].replace(np.nan,'UNKNOWN')
```

```python
df['Transaction Date']=df['Transaction Date'].ffill()
```

```python
df['Transaction Date']=pd.to_datetime(df['Transaction Date'])
```
